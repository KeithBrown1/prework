## A. match_ends

Given a list of strings, return the count of the number of strings where the string length is 2 or more and the first and last chars of the string are the same.

```python
def match_ends(words):
    count = 0
    for item in words:
        if len(item) >1:
            if item[:1]==item[-1:]:
                count += 1
    return count
```

```python
match_ends(['spam','eggs','spams','sketches'])
```

##B. front_x

Given a list of strings, return a list with the strings in sorted order, except group all the strings that begin with 'x' first.

e.g. ['mix', 'xyz', 'apple', 'xanadu', 'aardvark'] yields ['xanadu', 'xyz', 'aardvark', 'apple', 'mix']

Hint: this can be done by making 2 lists and sorting each of them before combining them


```python
def front_x(words):
    x_words =[]
    other_words =[]
    for item in words:
        if item[:1]=='x':
            x_words.append(item)
        else:
            other_words.append(item)
    x_words.sort()
    other_words.sort()
    combined = x_words + other_words
    return combined
```

```python
front_x(['mix', 'xyz', 'apple', 'xanadu', 'xavier', 'aardvark'])
```

##C. sort_last

Given a list of non-empty tuples, return a list sorted in increasing order by the last element in each tuple.

e.g. [(1, 7), (1, 3), (3, 4, 5), (2, 2)] yields [(2, 2), (1, 3), (3, 4, 5), (1, 7)]

Hint: use a custom key= function to extract the last element form each tuple.


```python
def sort_last(tuples):
    tuple_list =[]
    #defining index to sort by
    def last(item):
        index = item[-1]
        return index
    #sorting by index
    tuple_list = sorted(tuples, key=last)
    return tuple_list
```

```python
sort_last([(1, 7), (1, 3), (3, 4, 5), (2, 2)] )
```

##D. remove_adjacent

Given a list of numbers, return a list where all adjacent == elements have been reduced to a single element, so [1, 2, 2, 3] returns [1, 2, 3]. You may create a new list or modify the passed in list.


```python
def remove_adjacent(nums):
    length = len(nums)-1
    for item in xrange(length,0,-1):
        if nums[item] == nums[item-1]:
            del nums[item]
    return nums
```

```python
remove_adjacent([1, 2, 2, 3,3,3,3,3,4,4,5,6,1,'test','test'])
```

##E. linear_merge





Given two lists sorted in increasing order, create and return a merged list of all the elements in sorted order. You may modify the passed in lists. Ideally, the solution should work in "linear" time, making a single pass of both lists.


```python
def linear_merge(list1, list2):
    linear_sort = [] # output
    last_items = [] #remaing list items after sorting backwards
    empty = False #is a list empty
    while not empty:
        #selecting the last item from list1, and defining as a whilst the list has items
        #if the list is empty then, list 2 has the last sorted items to append
        if len(list1)>0:
            a = list1[-1]
        else:
            last_items = list2[:]
            empty = True
        
        #again for second list
        
        if len(list2)>0:
            b = list2[-1]
        else:
            last_items = list1[:]
            empty = True
            
        #Puting the largest availble on the end of the list, to give a sorted list
        #until one list runs out of items
        
        if not empty:
            if a > b:
                linear_sort.append(a)
                list1.pop()
            else:
                linear_sort.append(b)
                list2.pop()
        
    #reverse the last items as there may be more than one
    last_items.reverse()
        
    #combine sorted list and reversed remainder
    linear_sort += last_items
        
    #reversing the list to get back to increasing order
    linear_sort.reverse()
        
    return linear_sort
```

```python
linear_merge([1,2,3,4,7,8,9], [3,4,8])
```
